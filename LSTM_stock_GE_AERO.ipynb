```python
# Split data into training and testing sets (e.g., 80% train, 20% test)
train_size = int(len(df) * 0.8)
train_df, test_df = df[0:train_size], df[train_size:]

print("Shape of training data:", train_df.shape)
print("Shape of testing data:", test_df.shape)
train_df.head()
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense, Dropout
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
#from datetime import datetime

#Read the csv file
print(train_df.head()) #7 columns, including the Date.

#Separate dates for future plotting
train_dates = train_df['Date']
print(train_dates.tail(15)) #Check last few dates.

#Variables for training
cols = list(train_df)[1:6]
#Date and volume columns are not used in training.
print(cols) #['Open', 'High', 'Low', 'Close', 'Adj Close']

#New dataframe with only training data - 5 columns
df_for_training = train_df[cols].astype(float)

df_for_plot=df_for_training.tail(5000)
df_for_plot.plot.line()

#LSTM uses sigmoid and tanh that are sensitive to magnitude so values need to be normalized
# normalize the dataset
scaler = StandardScaler()
scaler = scaler.fit(df_for_training)
df_for_training_scaled = scaler.transform(df_for_training)


#As required for LSTM networks, we require to reshape an input data into n_samples x timesteps x n_features.
#In this example, the n_features is 5. We will make timesteps = 14 (past days data used for training).

#Empty lists to be populated using formatted training data
trainX = []
trainY = []

n_future = 1   # Number of days we want to look into the future based on the past days.
n_past = 14  # Number of past days we want to use to predict the future.

#Reformat input data into a shape: (n_samples x timesteps x n_features)
#In my example, my df_for_training_scaled has a shape (12823, 5)
#12823 refers to the number of data points and 5 refers to the columns (multi-variables).
for i in range(n_past, len(df_for_training_scaled) - n_future +1):
    trainX.append(df_for_training_scaled[i - n_past:i, 0:df_for_training.shape[1]])
    trainY.append(df_for_training_scaled[i + n_future - 1:i + n_future, 0])

trainX, trainY = np.array(trainX), np.array(trainY)

print('trainX shape == {}.'.format(trainX.shape))
print('trainY shape == {}.'.format(trainY.shape))

#In my case, trainX has a shape (12809, 14, 5).
#12809 because we are looking back 14 days (12823 - 14 = 12809).
#Remember that we cannot look back 14 days until we get to the 15th day.
#Also, trainY has a shape (12809, 1). Our model only predicts a single value, but
#it needs multiple variables (5 in my example) to make this prediction.
#This is why we can only predict a single day after our training, the day after where our data ends.
#To predict more days in future, we need all the 5 variables which we do not have.
#We need to predict all variables if we want to do that.

# define the Autoencoder model

model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
model.add(LSTM(32, activation='relu', return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(trainY.shape[1]))

model.compile(optimizer='adam', loss='mse')
model.summary()


# fit the model
history = model.fit(trainX, trainY, epochs=5, batch_size=16, validation_split=0.1, verbose=1)

plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.legend()

#Predicting...
#Libraries that will help us extract only business days in the US.
#Otherwise our dates would be wrong when we look back (or forward).
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay
us_bd = CustomBusinessDay(calendar=USFederalHolidayCalendar())
#Remember that we can only predict one day in future as our model needs 5 variables
#as inputs for prediction. We only have all 5 variables until the last day in our dataset.
n_past = 16
# n_days_for_prediction=15  #let us predict past 15 days

# Calculate the number of days to predict based on the desired date range
start_date = '2024-07-10'
end_date = '2025-07-10'
desired_forecast_dates = pd.date_range(start=start_date, end=end_date, freq=us_bd).tolist()

# Ensure the number of predictions doesn't exceed the available data points in trainX
n_days_for_prediction = min(len(desired_forecast_dates), len(trainX))

#Make prediction
prediction = model.predict(trainX[-n_days_for_prediction:]) #shape = (n, 1) where n is the n_days_for_prediction

#Perform inverse transformation to rescale back to original range
#Since we used 5 variables for transform, the inverse expects same dimensions
#Therefore, let us copy our values 5 times and discard them after inverse transform
prediction_copies = np.repeat(prediction, df_for_training.shape[1], axis=-1)
y_pred_future = scaler.inverse_transform(prediction_copies)[:,0]

# Generate forecast dates based on the number of predictions
last_train_date = train_dates.iloc[-1]
predict_period_dates = pd.date_range(start=last_train_date + us_bd, periods=n_days_for_prediction, freq=us_bd).tolist()


# Convert timestamp to date
forecast_dates = []
for time_i in predict_period_dates:
    forecast_dates.append(time_i.date())

print(f"Length of forecast_dates: {len(forecast_dates)}")
print(f"Length of y_pred_future: {len(y_pred_future)}")

df_forecast = pd.DataFrame({'Date':np.array(forecast_dates), 'Open':y_pred_future})
df_forecast['Date']=pd.to_datetime(df_forecast['Date'])


original = df[['Date', 'Open']].copy() # Create a copy to avoid SettingWithCopyWarning
original['Date']=pd.to_datetime(original['Date'])
original = original.loc[original['Date'] >= '2024-07-10']

```

# Task
Implement a backtesting strategy to evaluate the LSTM model's performance on the provided stock data.

## Define backtesting parameters

### Subtask:
Determine the size of the initial training window and the size of the testing (forecast) window for each backtesting iteration.


**Reasoning**:
Define the training and testing window sizes for the backtesting strategy based on the dataset size.



```python
# Determine the total size of the dataset
total_size = len(df)

# Define the size of the initial training window
# A common approach is to use a significant portion of the data for initial training
train_window_size = int(total_size * 0.8) # Using 80% of the data for initial training

# Define the size of the testing window (and the step size for rolling)
# This determines how many days are predicted in each iteration
test_window_size = 1 # Predicting one day at a time

print(f"Total dataset size: {total_size}")
print(f"Initial training window size: {train_window_size}")
print(f"Testing window size (and step size): {test_window_size}")
```

## Implement backtesting loop

### Subtask:
Iterate through the dataset, creating training and testing sets for each iteration based on the defined window sizes.


**Reasoning**:
Initialize lists for storing actual values and predictions, then loop through the dataset to create rolling training and testing windows and print their shapes.



```python
# Initialize lists to store actual values and predictions
all_actual_values = []
all_predictions = []

# Loop through the dataset, creating training and testing sets for each iteration
# The loop starts from the end of the initial training window
# up to the end of the dataset, with a step equal to the test_window_size.
for i in range(train_window_size, total_size, test_window_size):
    # Define the end index for the current training set
    train_end_index = i

    # Define the start and end indices for the current test set
    test_start_index = i
    test_end_index = i + test_window_size

    # Create the current training and testing DataFrames
    current_train_df = df.iloc[:train_end_index]
    current_test_df = df.iloc[test_start_index:test_end_index]

    # Print the shapes of the current training and testing DataFrames
    print(f"Iteration starting at index {i}:")
    print(f"  Shape of current_train_df: {current_train_df.shape}")
    print(f"  Shape of current_test_df: {current_test_df.shape}")
    print("-" * 20)
```

## Prepare data for lstm within loop

### Subtask:
For each iteration, scale and reshape the training and testing data for input into the LSTM model.


**Reasoning**:
Scale and reshape the training and testing data for input into the LSTM model as per the instructions.



```python
# Define the columns to be used for training and testing
cols_to_use = ['Open', 'High', 'Low', 'Close', 'Adj Close']

# Create new DataFrames with only the selected columns and convert to float
df_for_training = current_train_df[cols_to_use].astype(float)
df_for_test = current_test_df[cols_to_use].astype(float)

# Initialize a StandardScaler and fit it to the training data
scaler = StandardScaler()
scaler = scaler.fit(df_for_training)

# Transform both training and testing data
df_for_training_scaled = scaler.transform(df_for_training)
df_for_test_scaled = scaler.transform(df_for_test)

# Reshape training data for LSTM input
# n_past is defined in a previous cell
n_past = 14
n_features = df_for_training_scaled.shape[1] # Number of features (columns)

trainX = []
trainY = []

# The loop for creating trainX and trainY should go up to the last possible sequence end
# which is len(df_for_training_scaled) - n_future
n_future = 1 # Predicting one step ahead

for i in range(n_past, len(df_for_training_scaled) - n_future + 1):
    trainX.append(df_for_training_scaled[i - n_past:i, 0:n_features])
    trainY.append(df_for_training_scaled[i + n_future - 1:i + n_future, 0]) # Predicting the 'Open' price (index 0)

trainX, trainY = np.array(trainX), np.array(trainY)


# Reshape test data for LSTM input
# The testX sequences should end at the day *before* the testY value
# testY should be the actual values from the test set
testX = []
testY_actual = [] # Using a different name to avoid confusion with scaled trainY

# We need enough past data (n_past) before the first test data point
# to create the first sequence for prediction in the test set.
# This sequence will come from the end of the training data.
# The actual value to compare against will be the first value in the test set.

# For evaluating on the test set, we need sequences ending at index i-1 in the *combined* train+test data
# to predict the value at index i.
# Let's combine the scaled train and test data temporarily for sequence creation
combined_scaled_data = np.concatenate((df_for_training_scaled, df_for_test_scaled), axis=0)

# The test sequences start after the training data ends.
# The first test sequence ends at index train_end_index - 1 + n_past - 1 in the combined data
# The first actual value to predict is at index train_end_index in the original df
# So, the loop for creating testX and testY_actual should iterate through the test set indices
# and create sequences ending in the *combined* scaled data.

# The indices for the test set in the combined data start at train_end_index
# The actual values for testY_actual come from the original test_df
# The sequences for testX end at index i - 1 in the combined_scaled_data,
# where i is the index corresponding to the value being predicted in the combined_scaled_data.

# Let's simplify: testX will be sequences from the combined data that predict the values in the test set.
# The values to predict are the 'Open' prices in the current_test_df.
# The sequences to predict test_df.iloc[j] will end at index train_end_index + j - 1 in the combined_scaled_data.

# The number of test sequences we can create is len(df_for_test_scaled)
# For each index j in the test set (from 0 to len(df_for_test_scaled) - 1),
# the sequence for testX ends at index (train_end_index - 1) + j in the combined_scaled_data.
# The sequence starts at index (train_end_index - 1) + j - n_past + 1.

if len(df_for_test_scaled) > 0: # Only create test sequences if test set is not empty
    # We need at least n_past data points before the first prediction point in the test set.
    # These points come from the end of the training set.
    # So, the data needed for the first test sequence is from index train_end_index - n_past to train_end_index - 1
    # in the scaled training data.
    # And the value to predict is at index train_end_index in the original df (which is the first row of test_df).

    # Let's rethink the test data preparation for sequential prediction.
    # For backtesting, we are predicting test_window_size steps ahead based on the training data available up to train_end_index.
    # The test set `current_test_df` contains the actual values for this prediction period.
    # The input for predicting the first value in `current_test_df` should be the last `n_past` values from `current_train_df`.
    # Subsequent predictions within the test window would ideally use the predicted values (or a mix of actual and predicted)
    # as input, but for a simple evaluation, we can predict each step in the test window
    # using the last `n_past` *actual* values available up to that point.

    # However, the standard approach for evaluating an LSTM on a test set
    # is to create sequences from the test data itself (after the training data),
    # where each sequence predicts the next step within the test set.
    # This requires the test set to be at least `n_past` + `n_future` in length.
    # Given our test_window_size is 1, and n_past is 14, this approach might not be suitable
    # if test_window_size < n_past + n_future.

    # Let's use the approach where testX sequences are created from the combined scaled data,
    # predicting the corresponding values in the scaled test data.
    # This allows us to evaluate the model's performance on unseen data patterns.
    # The sequences for testX will be of length n_past.
    # The target values testY_actual will be the corresponding actual 'Open' prices from the test_df.

    # The number of test samples will be len(df_for_test_scaled).
    # For each sample j in the test set (from 0 to len(df_for_test_scaled)-1),
    # the sequence for testX will be the n_past values ending at index (train_end_index + j - 1) in the combined_scaled_data.
    # The value to predict is at index (train_end_index + j) in the combined_scaled_data, which corresponds to index j in df_for_test_scaled.

    # Let's create test sequences that end at the last available data point *before* the value we want to predict in the test set.
    # The first value we want to predict in the test set is df_for_test_scaled[0].
    # The sequence to predict this should end at index train_end_index - 1 in the scaled data.
    # So the first test sequence is combined_scaled_data[train_end_index - n_past : train_end_index]

    # The actual value for this first prediction is the first 'Open' price in the original test_df.
    # Subsequent predictions in the test set will use sequences ending within the test set itself.
    # For a test set of size `test_window_size`, we can make `test_window_size` predictions.
    # The input sequence for the j-th prediction (0-indexed) in the test set will be the `n_past` values ending at index `train_end_index + j - 1` in the combined scaled data.
    # The actual value for the j-th prediction is `current_test_df.iloc[j]['Open']`.

    # Let's create testX as sequences of length n_past ending at train_end_index - 1 + j for j from 0 to test_window_size - 1.
    # And testY_actual as the corresponding actual 'Open' values from current_test_df.

    testX = []
    testY_actual = []

    # The sequences for testX will be derived from the combined_scaled_data.
    # The indices in combined_scaled_data corresponding to the test set are from train_end_index to train_end_index + test_window_size - 1.
    # To predict the value at index k in combined_scaled_data (where k is in the test set range),
    # the input sequence is combined_scaled_data[k - n_past : k].

    # We want to predict each value in the test set.
    # For the first value in the test set (at index train_end_index in combined_scaled_data),
    # the sequence is combined_scaled_data[train_end_index - n_past : train_end_index].
    # For the second value (at index train_end_index + 1), the sequence is combined_scaled_data[train_end_index + 1 - n_past : train_end_index + 1].
    # ...
    # For the last value in the test set (at index train_end_index + test_window_size - 1),
    # the sequence is combined_scaled_data[train_end_index + test_window_size - 1 - n_past : train_end_index + test_window_size - 1].

    # This requires that train_end_index - n_past >= 0, which is true since train_end_index = i and i starts from train_window_size which is > n_past.

    for j in range(test_window_size):
        sequence_end_index_in_combined = train_end_index + j
        sequence_start_index_in_combined = sequence_end_index_in_combined - n_past
        testX.append(combined_scaled_data[sequence_start_index_in_combined:sequence_end_index_in_combined, 0:n_features])
        testY_actual.append(current_test_df.iloc[j]['Open']) # Append the actual Open price

    testX = np.array(testX)
    testY_actual = np.array(testY_actual)


    print('trainX shape == {}.'.format(trainX.shape))
    print('trainY shape == {}.'.format(trainY.shape))
    print('testX shape == {}.'.format(testX.shape))
    print('testY_actual shape == {}.'.format(testY_actual.shape))

else:
    print("Current test set is empty. Skipping data scaling and reshaping for this iteration.")
    print('trainX shape == {}.'.format(trainX.shape))
    print('trainY shape == {}.'.format(trainY.shape))
    print("testX and testY_actual are empty.")
```
